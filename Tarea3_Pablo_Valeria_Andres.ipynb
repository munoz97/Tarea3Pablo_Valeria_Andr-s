```python
# Código de solución estudiante 1 Pablo Muñoz
# Ejercicio 1
import numpy as np


def cong_method(D1:"Semilla inicial",D2:"Semilla inicial",n:"Número de elementos"):
    '''
    Esta función contiene la programación del método congruencial para la sucesión de la ecuación recursiva D_n = (n-1)D_(n-1) + (n-1)D_(n-1)
    '''
    D=np.zeros(n)
    D[0]=D1
    D[1]=D2
    def llenar_verctor(i):
        nonlocal D
        D[i]=((i+1)-1) * D[i-1] + ((i+1)-1 )* D[i-2]
    [llenar_verctor(i) for i in range(2,n)]  
    return D
result=cong_method(0,1,6)
result  
```

```python
##EJERCICIO 1
import numpy as np
def cong_method(d0:"Semilla1",d1:"Semilla2", N:"Número de elementos"):
    '''La función tiene el método congruencial para la sucesión D_n = ((n-1)D_{n−1} + (n-1)D_{n−2})
     para N>=3'''
    
    x = np.zeros(N)
    x[0] = d0
    x[1] = d1
    def vector(i):
        nonlocal x  
        x[i+2] = (N-1) * x[i+1] + (N-1) * x[i] 
    [vector(i) for i in range(N-2)]
    return x 

y = cong_method(0, 1, 3)
y
```
